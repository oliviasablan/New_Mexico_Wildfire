```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib as mplt
from matplotlib import rcParams
from matplotlib import rc
rc('mathtext', default='regular') 
rcParams['font.family'] = 'serif'
rcParams['mathtext.fontset'] = 'cm'
rcParams['mathtext.rm'] = 'serif'
rcParams['font.size'] = 12
```

```python
ED = pd.read_csv('../Data/health/ED/ED_data.csv')
ED['Date'] = pd.to_datetime(ED['Date'])
ED = ED[ED['Date'].dt.strftime('%B').isin(['April', 'May', 'June', 'July', 'August', 'September'])]
```

```python
races = ['American Indian/Alaskan Native', 'Asian/Native Hawaiian/Pacific Islander', 'Black', 'Hispanic', 'White', 'Unknown']
numbers = [1, 2, 3, 4, 5, 9]
print('Total Percentage of Race in ED data \n')
for i in range(len(races)):
    print( races[i], ': ', round(len(ED[ED['NMDOHRaceEth'] == numbers[i]])/len(ED)*100, 1))
```

```python
outcome = ["allresp1", "asthma1","COPD1","pneumonia1","bronchitis1", "allcardio1", "cardiacarrest1","arrythmia1","heartfail1","ischemic1","MI1","cerebrovascular1"]
titles = ['All Respiratory', 'Asthma', 'COPD', 'Pneumonia', 'Acute Bronchitis', 'All Cardiovascular', 'Cardiac Arrest', 'Arrhythmia', 'Heart Failure', 'Ischemic Heart Diesease', 
'Myocardial Infarction', 'Cerebrovacular Disease']

for i in range(len(outcome)):
    onediagnosis = ED[ED[outcome[i]] == 1]
    print('\n', titles[i], '\n')
    for ii in range(len(numbers)):
        print( races[ii], ': ', round(len(onediagnosis[onediagnosis['NMDOHRaceEth'] == numbers[ii]])/len(onediagnosis)*100, 1))
```

```python
print('Total Counts of Race in ED data \n')
print(len(ED))
for i in range(len(races)):

    print( races[i], ': ', len(ED[ED['NMDOHRaceEth'] == numbers[i]]))
```

### Repeat for Courtney's hospitalization data

```python
hosp_all = pd.read_csv("P:/EHEB/Climate Adaptation/HAQAST collaboration/Courtney/NMDOH_Data_for_CSU_full_details.csv")
hosp_all['Admission_Date'] = pd.to_datetime(hosp_all['Admission_Date'])
hosp_all = hosp_all[hosp_all['Admission_Date'].dt.strftime('%B').isin(['April', 'May', 'June', 'July', 'August', 'September'])]
```

```python
outcome = ['resp_prim_admits', 'asthma_prim_admits',  'copd_prim_admits','pneu_prim_admits', 'bron_prim_admits', 'cardio_prim_admits', 
       'cardiac_prim_admits', 'arrythmia_prim_admits','heartfail_prim_admits', 'ischemic_prim_admits','mi_prim_admits','cerebro_prim_admits']
titles = ['All Respiratory', 'Asthma', 'COPD', 'Pneumonia', 'Acute Bronchitis', 'All Cardiovascular', 'Cardiac Arrest', 'Arrhythmia', 'Heart Failure', 'Ischemic Heart Diesease', 
'Myocardial Infarction', 'Cerebrovacular Disease']

for i in range(len(outcome)):
    onediagnosis = hosp_all[hosp_all[outcome[i]] == 1]
    print('\n', titles[i], '\n')
    for ii in range(len(numbers)):
        if ii < 5:
            print( races[ii], ': ', round(len(onediagnosis[onediagnosis['NMDOHRaceEth'] == numbers[ii]])/len(onediagnosis)*100, 1))
        else:
            print( races[ii], ': ', round(len(onediagnosis[~onediagnosis['NMDOHRaceEth'].isin(numbers)])/len(onediagnosis)*100, 1))
```

```python
print('Total Counts of Race in Hospitalization data (', str(len(hosp_all)), ')\n')
for i in range(len(races)):
    if i < 5 :
        print( races[i], ': ', len(hosp_all[hosp_all['NMDOHRaceEth'] == numbers[i]]))
    else: 
        print( races[i], ': ', len(hosp_all[~hosp_all['NMDOHRaceEth'].isin(numbers)]))
```

```python
years = ['2016', '2017', '2018', '2019', '2020', '2021', '2022']
for i in range(len(years)):
    year = hosp_all[hosp_all['Admission_Date'].dt.strftime('%Y') == years[i]]
    print('\n', years[i], '\n')
    for ii in range(len(numbers)):
        if ii < 5:
            print( races[ii], ': ', round(len(year[year['NMDOHRaceEth'] == numbers[ii]])/len(year)*100, 1))
        else:
            print( races[ii], ': ', round(len(year[~year['NMDOHRaceEth'].isin(numbers)])/len(year)*100, 1))
```

```python
print('Total Percentage of Race in hosp data \n')
for ii in range(len(numbers)):
    if ii < 5:
        print( races[ii], ': ', round(len(hosp_all[hosp_all['NMDOHRaceEth'] == numbers[ii]])/len(hosp_all)*100, 1))
    else:
        print( races[ii], ': ', round(len(hosp_all[~hosp_all['NMDOHRaceEth'].isin(numbers)])/len(hosp_all)*100, 1))
```
