Data From: https://gacc.nifc.gov/swcc/predictive/intelligence/Historical/Fire_and_Resource_Data/Historical_Fires_Acres.htm

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib as mplt
from matplotlib import rcParams
from matplotlib import rc
rc('mathtext', default='regular') 
rcParams['font.family'] = 'serif'
rcParams['mathtext.fontset'] = 'cm'
rcParams['mathtext.rm'] = 'serif'
rcParams['font.size'] = 12
```

```python
acresall = pd.read_csv('../Data/smoke/All_Fires_Average_Per_Year.csv')
acres = acresall[acresall['YEAR'] > 2015]
```

```python
fig, ax = plt.subplots(figsize = (8, 6))
ax2 = ax.twinx()

ax.bar(acres['YEAR'], acres['ACRES']/1000000, color = 'xkcd:cloudy blue', edgecolor = 'k')
ax.set_xlim([2015, 2024])

ax.set_xticks(np.arange(2016,2024, 1));
ax.set_xticklabels(ax.get_xticklabels(), rotation=45)

ax.set_ylabel('Millions of Acres Burned')

ax2.scatter(acres['YEAR'], acres['FIRES'], color = 'xkcd:ultramarine')
ax2.set_ylabel('Number of Fires')
plt.xticks(rotation=45);
ax2.spines['right'].set_visible(False)
ax2.spines['top'].set_visible(False)
```

```python
fig, ax = plt.subplots(figsize = (8, 6))
ax.bar(acres['YEAR'], acres['ACRES']/1000000, color = 'xkcd:cloudy blue', edgecolor = 'k')
ax.set_xlim([2015, 2024])
ax.set_xticks(np.arange(2016,2024, 1));
ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
ax.set_ylabel('Millions of Acres Burned')
ax2.spines['right'].set_visible(False)
ax2.spines['top'].set_visible(False)
plt.axhline(y = np.mean(acres['ACRES']/1000000), color = 'xkcd:ultramarine', label = 'Average')
ax.legend(fontsize = 10, frameon = False, bbox_to_anchor = (.82, 0.31))
```
